```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
ansurm = pd.read_csv('../Datasets/ansur_male.csv')
ansurm.head()
```

```python
ansurm.drop('Gender', axis=1, inplace=True)
```

```python
ansurm.dropna(inplace=True)
```

```python
ansurm_cat = ansurm[ansurm.columns[ansurm.dtypes==object]]
```

```python
ansurm_num= ansurm[ansurm.columns[~(ansurm.dtypes==object)]]
```

```python
X = ansurm_num
```

```python
y= ansurm_cat['Height_class'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25)
```

```python
scaler=StandardScaler()
```

```python
X_trains = scaler.fit_transform(X_train)
```

```python
X_tests = scaler.transform(X_test)
```

```python
from sklearn.decomposition import PCA
```

```python
pca= PCA()
```

```python
pc = pca.fit_transform(X_trains)
```

```python
pca.explained_variance_ratio_.cumsum()
```

```python
pca.components_[4]
```

```python
pc2 = pc[:,:2]
```

```python
pc3 = pd.DataFrame(pc2, columns=['PC1', 'PC2'])
```

```python
ansurm_cat.columns
```

```python
pc3[ansurm_cat.columns]= ansurm_cat
```

```python
pc3.head()
```

```python
for f in ansurm_cat.columns:
    sns.scatterplot(data=pc3, x='PC1', y='PC2', hue=f)
    plt.show()
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
pipe = Pipeline([('scaler', StandardScaler()), ('PCA', PCA(n_components=10)), ('classifier', DecisionTreeClassifier()) ])
```

```python
pipe.fit(X_train, y_train)
```

```python
pipe['PCA'].explained_variance_ratio_.sum()
```

```python
pipe.score(X_test,y_test)
```

```python
components= pipe['PCA'].components_
```

```python
str(dict(zip(ansurm_num.columns, components[0])))
```

```python
ratio = pipe['PCA'].explained_variance_ratio_
```

```python
plt.plot(ratio)
plt.xlabel('Components')
plt.ylabel('Variance')
```
